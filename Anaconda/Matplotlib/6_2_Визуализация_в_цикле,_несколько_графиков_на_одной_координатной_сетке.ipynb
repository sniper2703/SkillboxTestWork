# Продвинутая визуализация с matplotlib

## Урок 2. Визуализация в цикле, несколько графиков на одной координатной сетке

В этом уроке мы научимся изображать несколько графиков на одной координатной сетке, а также выводить графики в цикл.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Разбор практического задания

```python
df = pd.read_csv('wage-data-coast.csv')
df
```

```python
data = df[(df['State'] == 'Arizona') & (df['Year'] <= 2000)][['Year', 'Salary']]

fig, ax = plt.subplots()
ax.plot(data['Year'], data['Salary'])
ax.set_title('Минимальная зарплата в штате Аризона по годам', color='navy')
ax.set_xlabel('Год', color='navy')
ax.set_ylabel('Минимальная зарплата, $/час', color='navy')

plt.show()
```

#### **Вывод нескольких графиков на одной координатной сетке**


Из графика выше кажется, что в 1999 году произошел достаточно сильный рост минимальной зарплаты, и после этого года зарплата росла почти каждый год, и мы хотим это показать. Давайте вертикальной линией разделим график на 2 периода - до 1999 года и после.

Мы хотим добавить вертикальную линию к уже существующему графику, который описывается объектом ax. Для построения вертикальной линии используем метод `ax.axvline()`, в который передадим параметр x=199.


```python
fig, ax = plt.subplots()

data = df[df['State'] == 'Washington'][['Year', 'Salary']]

ax.plot(data['Year'], data['Salary'])
ax.axvline(x = 1999)

ax.set_title('Минимальная зарплата в штате Вашингтон по годам', pad = 16, color = 'navy', backgroundcolor = 'lightgray')
ax.set_xlabel('Год')
ax.set_ylabel('Минимальная зарплата, $/час')
ax.set_xticks(list(range(data['Year'].min(), data['Year'].max(),5))+[data['Year'].max()])

plt.show()
```

Итак, линию мы добавили, но теперь было бы неплохо выделить ее другим цветом, например, зеленым. Для этого в метод `axvline()` добавим параметр **color** и присвоим ему значение green, т.е. зеленый.

```python
fig, ax = plt.subplots()

data = df[df['State'] == 'Washington'][['Year', 'Salary']]

ax.plot(data['Year'], data['Salary'])
ax.axvline(x = 1999, color = 'green')

ax.set_title('Минимальная зарплата в штате Вашингтон по годам', pad = 16, color = 'navy', backgroundcolor = 'lightgray')
ax.set_xlabel('Год')
ax.set_ylabel('Минимальная зарплата, $/час')
ax.set_xticks(list(range(data['Year'].min(), data['Year'].max(),5))+[data['Year'].max()])

plt.show()
```

Выглядит неплохо, но может быть, эту линию еще можно как-то отформатировать? Например, сделать тоньше с помощью изменения параметра **linewidh**.

```python
fig, ax = plt.subplots()

data = df[df['State'] == 'Washington'][['Year', 'Salary']]

ax.plot(data['Year'], data['Salary'])
ax.axvline(x = 1999, color = 'green', linewidth = 1)

ax.set_title('Минимальная зарплата в штате Вашингтон по годам', pad = 16, color = 'navy', backgroundcolor = 'lightgray')
ax.set_xlabel('Год')
ax.set_ylabel('Минимальная зарплата, $/час')
ax.set_xticks(list(range(data['Year'].min(), data['Year'].max(),5))+[data['Year'].max()])

plt.show()
```

Получили тонкую зеленую линию, как мы и хотели. Давайте попробуем сделать ее прерывистой. Для этого есть параметр **linestyle**, который принимает значение дефиса, двух дефисов и так далее. Попробуем 2 дефиса и посмотрим, что получится.


```python
fig, ax = plt.subplots()

data = df[df['State'] == 'Washington'][['Year', 'Salary']]

ax.plot(data['Year'], data['Salary'])
ax.axvline(x = 1999, color = 'green', linewidth = 1, linestyle = '--')

ax.set_title('Минимальная зарплата в штате Вашингтон по годам', pad = 16, color = 'navy', backgroundcolor = 'lightgray')
ax.set_xlabel('Год')
ax.set_ylabel('Минимальная зарплата, $/час')
ax.set_xticks(list(range(data['Year'].min(), data['Year'].max(),5))+[data['Year'].max()])

plt.show()
```

Теперь давайте на этих же осях отобразим график минимальной зарплаты в Калифорнии. Для этого создадим еще один датафрейм, только вместо штата Вашингтон выберем данные штата Калифорния. 

Построим еще один линейный график. Поскольку мы хотим остаться в этих же координатных осях, то вызовем метод `plot()` у того же объекта ax. В метод `plot()` передадим массив годов и массив зарплат.

```python
fig, ax = plt.subplots()

data = df[df['State'] == 'Washington'][['Year', 'Salary']]
data1 = df[df['State'] == 'California'][['Year', 'Salary']]

ax.plot(data['Year'], data['Salary'])
ax.plot(data1['Year'], data1['Salary'])
ax.axvline(x = 1999, color = 'green', linewidth = 1, linestyle = '--')

ax.set_title('Минимальная зарплата в штате Вашингтон по годам', pad = 16, color = 'navy', backgroundcolor = 'lightgray')
ax.set_xlabel('Год')
ax.set_ylabel('Минимальная зарплата, $/час')
ax.set_xticks(list(range(data['Year'].min(), data['Year'].max(),5))+[data['Year'].max()])

plt.show()
```

Теперь на нашем графике фактически размещаются 3 графика. И если вертикальную линию мы не спутаем с другими, то какая из линий соответствует штату Вашингтон, а какая - Калифорнии, не совсем понятно. Значит, нужно добавить легенду. Сделать это можно с помощью метода ax.legend().

```python
fig, ax = plt.subplots()

data = df[df['State'] == 'Washington'][['Year', 'Salary']]
data1 = df[df['State'] == 'California'][['Year', 'Salary']]

ax.plot(data['Year'], data['Salary'])
ax.plot(data1['Year'], data1['Salary'])
ax.axvline(x = 1999, color = 'green', linewidth = 1, linestyle = '--')

ax.set_title('Минимальная зарплата в штате Вашингтон по годам', pad = 16, color = 'navy', backgroundcolor = 'lightgray')
ax.set_xlabel('Год')
ax.set_ylabel('Минимальная зарплата, $/час')
ax.set_xticks(list(range(data['Year'].min(), data['Year'].max(),5))+[data['Year'].max()])
ax.legend()

plt.show()
```

Текстом `matplotlib` говорит нам, что не все в порядке. Для того, чтобы вывести легенду на график, нужно, чтобы хотя бы для одной линии на графике была указана ее метка, т.е. лейбл. Давайте добавим лейблы линейным графикам.

```python
fig, ax = plt.subplots()

data = df[df['State'] == 'Washington'][['Year', 'Salary']]
data1 = df[df['State'] == 'California'][['Year', 'Salary']]

ax.plot(data['Year'], data['Salary'], label = 'Washington')
ax.plot(data1['Year'], data1['Salary'], label = 'California')
ax.axvline(x = 1999, color = 'green', linewidth = 1, linestyle = '--')

ax.set_title('Минимальная зарплата в штате Вашингтон по годам', pad = 16, color = 'navy', backgroundcolor = 'lightgray')
ax.set_xlabel('Год')
ax.set_ylabel('Минимальная зарплата, $/час')
ax.set_xticks(list(range(data['Year'].min(), data['Year'].max(),5))+[data['Year'].max()])
ax.legend()

plt.show()
```

Видим, что легенда появилась.

Давайте теперь построим новый график с несколькими гистограммами. Допустим, мы хотим сравнить распределения минимальных зарплат в 2000 и 2010 годах.

Сначала нам нужно выбрать данные. Давайте сначала выберем данные за 2000 год, и оставим только признак **Salary**. Сделаем то же самое, только для 2010 года.


```python
data = df[df['Year'] == 2000]['Salary']
data1 = df[df['Year'] == 2010]['Salary']
```

Теперь переходим к инициализации графика. Давайте создадим график размером 12 на 6. Затем добавим гистограммы на график. Для этого у объекта ax есть метод `hist`, который принимает один массив чисел. Добавим массив зарплат за 2000 год. И давайте сразу укажем лейбл, чтобы потом добавить легенду на график. Теперь на этот же график добавим вторую гистограмму, т.е. у этого же объекта ax вызовем метод `hist` еще раз. Передадим массив зарплат за 2010 год, и укажем лейбл.

Теперь добавим заголовок, подписи осей и легенд.

```python
fig, ax = plt.subplots(figsize=(12,6))
ax.hist(data, label = '2000')
ax.hist(data1, label = '2010')

ax.set_title('Распределение минимальных зарплат в 2000 и 2010 годах')
ax.set_xlabel('Минимальная зарплата, $/час')
ax.set_ylabel('# записей')
ax.legend()

plt.show()
```

В таком виде гистограмму читать не совсем удобно, потому что за оранжевой областью справа могут быть синие столбцы, но мы их сейчас не видим. Давайте сделаем заливку столбцов не сплошной, как сейчас, а немного прозрачной, чтобы видеть перекрывающиеся области. 

Для этого у метода `hist` есть параметр **alpha**, который означает степень прозрачности. **Alpha** принимает значения от 0 до 1, 0 - это полностью прозрачная заливка, а 1 - сплошная, как сейчас. Давайте у обоих графиков зададим параметр **alpha** равным 0.5.


```python
fig, ax = plt.subplots(figsize=(12,6))
ax.hist(data, label = '2000', alpha = 0.5)
ax.hist(data1, label = '2010', alpha = 0.5)

ax.set_title('Распределение минимальных зарплат в 2000 и 2010 годах')
ax.set_xlabel('Минимальная зарплата, $/час')
ax.set_ylabel('# записей')
ax.legend()

plt.show()
```

Давайте теперь посмотрим на ось Ох. Текущие отсечки кажутся неудобными, потому что непонятно, где граница каждого столбца. 

Метод `hist` возвращает 3 объекта, и второй - **bins**, - содержит отсечки для горизонтальной оси.

Мы можем получить отсечки каждой гистограммы по отдельности, и нанести на ось Ох отсечки обеих гистограмм. Давайте так и сделаем сейчас. Нам нужно записать отсечки в отдельные переменные. `hist` возвращает три объекта, при этом нам не нужен первый и третий объект. Их мы запишем в анонимные переменные, а второй объект, **bins**, в соответствующую переменную. 



```python
fig, ax = plt.subplots(figsize=(12,6))
_, bins, _ = ax.hist(data, label = '2000', alpha = 0.5)
_, bins1, _ = ax.hist(data1, label = '2010', alpha = 0.5)
```

Теперь нанесем оба набора отсечек на ось с помощью метода `set_xticks`. Приведем оба списка к типу `list` и просто сложим их.

```python
fig, ax = plt.subplots(figsize=(12,6))
_, bins, _ = ax.hist(data, label = '2000', alpha = 0.5)
_, bins1, _ = ax.hist(data1, label = '2010', alpha = 0.5)

ax.set_title('Распределение минимальных зарплат в 2000 и 2010 годах')
ax.set_xlabel('Минимальная зарплата, $/час')
ax.set_ylabel('# записей')
ax.set_xticks(list(bins) + list(bins1))
ax.legend()

plt.show()
```

Для того, чтобы установить угол поворота отсечек на оси Ох, нужно вызвать метод `ax.tick_params()`. В этот метод мы передадим ось, в нашем случае х, и укажем, какой атрибут этой оси мы хотим установить. За поворот отвечает параметр **rotation**, и давайте повернем отсечки на 45 градусов.

```python
fig, ax = plt.subplots(figsize=(12,6))
_, bins, _ = ax.hist(data, label = '2000', alpha = 0.5)
_, bins1, _ = ax.hist(data1, label = '2010', alpha = 0.5)

ax.set_title('Распределение минимальных зарплат в 2000 и 2010 годах')
ax.set_xlabel('Минимальная зарплата, $/час')
ax.set_ylabel('# записей')
ax.set_xticks(list(bins) + list(bins1))
ax.tick_params(axis = 'x', rotation = 45)
ax.legend()

plt.show()
```

Стало немного лучше, но некоторые отсечки все равно нечитаемы. Чтож, давайте перестроим гистограммы так, чтобы ширина столбцов у обоих графиков была одинакова. Для этого мы сначала построим более широкую гистограмму, то есть гистограмму за 2010 год, после этого сохраним в отдельную переменную ширину всех столбцов, и построим вторую гистограмму со столбцами такой же ширины. Давайте скопируем код в новую ячейку и удалим строку с поворотом текста отсечек. 

Теперь поменяем порядок построения гистограмм. Сначала построим более широкую гистограмму для 2010 года и сохраним отсечки в переменную. Теперь строим вторую гистограмму, и передаем сохраненные отсечки в параметр **bins**. Таким образом, столбцы второй гистограммы будут такой же ширины, что и первой. Теперь осталось передать эти же отсечки в метод `set_xticks`, чтобы отрисовать их на оси Ох.


```python
fig, ax = plt.subplots(figsize=(12,6))
_, bins, _ = ax.hist(data1, label = '2010', alpha = 0.5)
ax.hist(data, label = '2000', alpha = 0.5)

ax.set_title('Распределение минимальных зарплат в 2000 и 2010 годах')
ax.set_xlabel('Минимальная зарплата, $/час')
ax.set_ylabel('# записей')
ax.set_xticks(list(bins))
ax.tick_params(axis = 'x', rotation = 45)
ax.legend()

plt.show()
```

Давайте уточним гистограммы, сделав столбцы yже, т.е. увеличим количество столбцов на графике. Для этого в первый график в параметр **bins** передадим число 20 - это значит, что на гистограмме должно быть 20 столбцов. 

```python
fig, ax = plt.subplots(figsize=(12,6))
_, bins, _ = ax.hist(data1, label = '2010', alpha = 0.5, bins = 20)
ax.hist(data, label = '2000', alpha = 0.5, bins = bins)

ax.set_title('Распределение минимальных зарплат в 2000 и 2010 годах')
ax.set_xlabel('Минимальная зарплата, $/час')
ax.set_ylabel('# записей')
ax.set_xticks(list(bins))
ax.tick_params(axis = 'x', rotation = 45)
ax.legend()

plt.show()
```

#### **Вывод графиков в цикле**

В некоторых случаях выводить несколько графиков в одной координатной сетке не совсем удобно, например, если нам нужно посмотреть одновременно не 2, а 5 гистограмм. В таких случаях уместно выводить графики, например, друг под другом, чтобы иметь возможность сравнить их глазами.

Скопируем код в новую ячейку и немного изменим его. Первой строкой добавим цикл. Мы хотим построить гистограммы для 2000 и 2010 года, по ним и будем итерировать. Сдвинем весь код в тело цикла. Теперь удалим построение второго графика, потому что мы будем выводить только один график за раз. Заменим явный год на переменную year. Также обновим лейбл. Прозрачность нам не нужна, поэтому параметр **alpha** удалим. Заголовок мы удалим, и легенда нам, в принципе, тоже не нужна.

```python
for year in [2000, 2010]:
  data = df[df["Year"] == year]['Salary']

  fig, ax = plt.subplots(figsize=(12,6))
  _, bins, _ = ax.hist(data, label = str(year), bins = 20)
  
  ax.set_title('Распределение минимальных зарплат в ' + str(year) + ' году')
  ax.set_xlabel('Минимальная зарплата, $/час')
  ax.set_ylabel('# записей')
  ax.set_xticks(list(bins))
  ax.tick_params(axis = 'x', rotation = 45)
```
