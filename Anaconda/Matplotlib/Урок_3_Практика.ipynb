Обновите график среднего значения минимальной зарплаты, который вы создали на предыдущем практическом задании. Добавьте к нему соответствующий заголовок, названия осей и обновите метки горизонтальной оси.

```python
df = pd.read_csv('wage-data.csv')
df.head()
```

```python
series = df.groupby('Year')['Salary'].mean()
series
```

```python
x_range = list(range(series.index.min(), series.index.max()+1, 10))
x_range.append(series.index.max())
```

```python
plt.plot(series)
plt.title("Динамика средней зарплаты по всем штатам по годам")
plt.xlabel('Год')
plt.ylabel('Минимальная зарплата, $/час')
plt.xticks(x_range)
plt.xticks(x_range, rotation=45)
```
